```python
#converting the unit in the height column to cm
```

```python
def convert_height(x):
    if x[-1] == '"':
        x = x.replace("\"","")
        inch = int(x[2:]) * 2.54
        foot = int(x[0]) * 30.48
        return round(foot+inch)
    elif x[-1] == "m":
        return int(x[:-2])
```

```python
df.Height = df.Height.apply(convert_height)
```

```python
df.Height.unique()
```

```python
#shall we do the same for the weight column
```

```python
def convert_weight(x):
    if x[-1] == 's':
        x = x.replace("lbs","")
        kg = int(x[:3]) / 2.205
        return round(kg)
    elif x[-1] == "g":
        return int(x[:-2])
```

```python
df.Weight = df.Weight.apply(convert_weight)
```

```python
# As we can see the values of the variables are represented in 'M' and 'k' which means millions and thousand 
#let's see if we can represent them in integers. shall we...
```

```python
def release(Y):
    if 'M' in Y:
        Y = Y.replace('M', '')
        Y = float(Y[1:]) * 1000000
        return round(Y)
    elif "K" in Y:
        Y = Y.replace('K', '')
        Y = float(Y[1:]) * 1000
        return round(Y)
    else:
        return Y
```

```python
df["Release Clause"] = df["Release Clause"].apply(release)
```

```python
df["Release Clause"]
```

```python
#lets treat the wage column as well
```

```python
def wage(x):
    if 'K' in x:
        x = x.replace('K', '')
        x = int(x[1:]) * 1000
        return round(x)
    else:
        return int(x[1:])
```

```python
df.Wage = df.Wage.apply(wage)
```

```python
df.Wage
```

```python
# repeating the same process lets treat the value column as well
```

```python
def value(x):
    if 'M' in x:
        x = x.replace('M', '')
        x = float(x[1:]) * 1000000
        return round(x)
    elif "K" in x:
        x = x.replace('K', '')
        x = float(x[1:]) * 1000
        return round(x)
    else:
        return int(x[1:])
```

```python
df.Value = df.Value.apply(value)
```

```python
''''The next is the contract column. lets work on creating categorical variables while assigning the variables
to a new column
```

```python
def check_contract(val):
    if 'On Loan' in val:
        return'Loanee'
    elif 'Free' in val:
        return 'Free Agent'
    elif '~' in val:
        return 'Under Contract'
    else:
        return 'Unknown'
```

```python
df['contract_type'] = df['Contract'].apply(lambda x : check_contract(x)) 
```

```python
df.contract_type.value_counts()
```

```python
#create a new column for contract_started and contract_ends
```

```python
# extract year from last 4 characters of 'Joined' column
df['Contract_start'] = df['Joined'].apply(lambda x: x[-4:])
```

```python
def value(val):
    if 'Loan' in val:
        val = val[7:12]
        val = val.strip(' ') 
        return val
    else:
        val = val[-4:]
        return val
        
df['Contract_end'] = df['Contract'].apply(lambda x : value(x)) 
```

```python
''''i noticed that some attributes such as attacking, Defending, Skill, Power, Defending, Mentality, Movement
and Goalkeeping are the sum of other stats under them. lets calculate the % for the attributes with the total
```

# Attacking = Crossing + Heading Accuracy + Finishing + Short Passing + Short Passing + Volleys

```python
def attacking(val):
    val = (val/500) * 100
    return val

df['Attacking_percentage'] = df['Attacking'].apply(lambda x: attacking(x))
```

# Skill = Dribbling + FK Accuracy + Curve + Long Passing + Ball Control

```python
def skill(val):
    val = (val/500) * 100
    return val

df['Skill_percentage'] = df['Skill'].apply(lambda x: skill(x))
```

# Power = Shot Power + Jumping + Stamina + Strength + Long Shots

```python
def power(val):
    val = (val/500) * 100
    return val

df['Power_percentage'] = df['Power'].apply(lambda x: power(x))
```

# Defending = Standing Tackle + Sliding Tackle + Marking

```python
def defending(val):
    val = round(((val/300) * 100),2)
    return val

df['Defending_percentage'] = df['Defending'].apply(lambda x: defending(x))
```

# Mentality = Aggression + Interceptions + Positioning + Vision + Penalties + Composure

```python
def mentality(val):
    val = round(((val/600) * 100),2)
    return val

df['Mentality_percentage'] = df['Mentality'].apply(lambda x: mentality(x))
```

# Movement = Acceleration + Sprint Speed + Agility + Reactions + Balance

```python

def movement(val):
    val = (val/500) * 100
    return val

df['Movement_percentage'] = df['Movement'].apply(lambda x: movement(x))
```

# Goalkeeping = GK Diving + GK Handling + GK Kicking + GK Positioning + GK Reflexes 

```python
def goalkeeping(val):
    val = (val/500) * 100
    return val

df['GoalKeeping_percentage'] = df['Goalkeeping'].apply(lambda x: goalkeeping(x))
```

# Total Stats = Attacking + Skill + Power + Defending + Mentality + Movement + Goalkeeping

```python
def totalstats(val):
    val = (val/700) * 100
    return val

df['totalstats_percentage'] = df['Total Stats'].apply(lambda x: totalstats(x))
```
